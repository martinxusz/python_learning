```python
# -*- coding: utf-8 -*-
#!/usr/bin/python
import requests
import json


foo = {
    'accId':'607283',
'm':'IPhone6',
'classId':'7511',
'imei':'20161128193454GBNAIVZLZRKHB',
'msgType':'9',
'ov':'4.4.2',
'pageCount':'20',
'pageNo':'1',
'plat':'2',
'schoolId':'1482',
'sid':'6072831488114861907',
'src':'appstore',
't':'1488102328852',
'userId':'690001',
'ver':'4.5.0'
}



def retrive_page(url, json_obj):
    headers = {'Content-type': 'application/json', 'Accept': 'text/plain'}
    r = requests.post(url, data=json_obj, headers=headers)
    if r.status_code == 200:
        #取出结果开始分析
        result_json = r.json()
        
        #如果没有内容就不解析了
        if result_json['shareList'] == None or len(result_json['shareList']) == 0:
            return False
        else:
            #否则遍历结果中的每一条消息
            for nodeitem in result_json['shareList']:
                #只抓带有图片内容的，无图无真相的就跳过不要吧
                if nodeitem['imgList'] != None and len(nodeitem['imgList']) > 0:
                    for eachimg in nodeitem['imgList']:
                        print eachimg
                if nodeitem['videoUrl'] != None and len(nodeitem['videoUrl']) > 0 :
                        #如果只含一个视频（字符串），就直接打出
                        if isinstance(nodeitem['videoUrl'], basestring):
                            print nodeitem['videoUrl']
                        else:
                            #否则为列表，需要挨个打出
                            for eachvideo in nodeitem['videoUrl']:
                                    print eachvideo
        return True
    else:
        print 'error with code:' + str(r.status_code)
        return False

```

```python
json_foo = json.dumps(foo)
url = 'http://api.yzmm365.cn/gardenapp/garden/queryMsg'


while retrive_page(url, json_foo) is True:
    #继续抓取下一页
    foo['pageNo'] = str( int(foo['pageNo']) + 1 )
    json_foo = json.dumps(foo)
    print '------------------- Next Page-----------------------------------'

#下一步计划将这个URL下载保存到硬盘上
print '#下一步计划将这个URL下载保存到硬盘上'
```
