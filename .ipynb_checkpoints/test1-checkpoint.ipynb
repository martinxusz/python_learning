取出某日涨停的所有股票

```python
import tushare as ts
from tushare.util import dateu as du
#一次性获取当前交易所有股票的行情数据（如果是节假日，即为上一交易日，结果显示速度取决于网速）
#df = ts.get_today_all()
#全部历史数据，请调用下一个接口get_h_data()

#某支股票某日的数据
df=ts.get_h_data('601872',start=du.last_tddate())
df
```

```python
#找出过去60天的数据
df1=ts.get_h_data('601872',start=du.day_last_week(-60))
df1
```

```python
#过滤出下跌的记录
df1[(df1.open > df1.close)]
```

```python
%matplotlib inline
df1.plot()
```
